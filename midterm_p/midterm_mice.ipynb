```python
import pylab
import numpy as np
import math

def generate_ring(r,N):
    
    x = []
    y = []
    
    for i in range(N):
        x.append(r*math.cos((2*math.pi*i)/N))
        y.append(r*math.sin((2*math.pi*i)/N))
    return x,y

def generate_mice(r):
    fig, axes = pylab.subplots()
    for i in range(12):
        x , y = generate_ring((0.2 + 0.05*i)*r, (12 + 3*i))
        axes.scatter(x,y, fc='b', s = 3)
    axes.scatter(0,0, fc = 'b', s = 3)
    pylab.savefig("mice.png")
```

```python
generate_mice(100)
```
